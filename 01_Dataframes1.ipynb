# Column operations

```python
df['total_bill']
```

```python
type(df['total_bill'])
```

```python
mylist = ['total_bill', 'price_per_person']
df[mylist]
```

```python
df['total_bill'] + df['tip']
```

```python
# tip percentage
df['tip_percentage'] = 100 * df['tip'] / df['total_bill']
```

```python
df.head()
```

```python
df
```

```python
df['tip_percentage'] = np.round(df['tip_percentage'], 2)
```

```python
df.head()
```

```python
# Droping or deleting a column
df.drop('tip_percentage', axis=1)
```

```python
df.head()
```

**But this change is not permanent**
**To  make it permanent use this approach**

```python
df = df.drop('tip_percentage', axis=1)
```
