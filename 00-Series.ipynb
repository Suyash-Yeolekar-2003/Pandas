# Series
A Series a data structure in python that holds an array of information along with named index.

The named index differentiates it from the simple numpy array.

## Formal definition: 
One- dimensional ndarray with axis labels 

```python
import numpy as np
import pandas as pd
```

```python
mylist = ['USA', 'CANADA', 'MEXICO']
```

```python
mydata = [1776, 1867, 1821]
```

```python
myser = pd.Series(data = mydata, index=mylist)
```

```python
myser
```

```python
myser[0]
```

```python
myser['USA']
```

### We can also give dictionary as a parameter to Series.

Pandas is smart enough that it can automatically keep dict key as Series label and value as data.

```python
ages = {'Sam': 10, 'Rohan': 21, 'Sandesh': 25}
```

```python
series1 = pd.Series(ages)
series1
```

```python
# Imaginary Sales Data for 1st and 2nd Quarters for Global Company
q1 = {'Japan': 80, 'China': 450, 'India': 200, 'USA': 250}
q2 = {'Brazil': 100,'China': 500, 'India': 210,'USA': 260}
```
