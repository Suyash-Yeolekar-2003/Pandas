**In case you need to get only last 4 digits of CC Number, then we'll do it using .apply method**

The apply() method allows you to apply a function along one of the axis of the DataFrame, default 0, which is the index (row) axis.

```python
def last_four(num):
  return int(str(num)[-4:])
```

```python
last_four(123456789)
```

```python
type(last_four(123456789))
```

**Applying it on the df **

```python
df['CC Number'].apply(last_four)
```

```python
df['Last_four'] = df['CC Number'].apply(last_four)
df.head()
```

```python
def yelp(price):
  if price < 10:
    return '$'
  elif price < 30:
    return '$$'
  else:
    return '$$$'
```

```python
yelp(25)
```

```python
df['Yelp'] = df['total_bill'].apply(yelp)
```

```python
df
```

```python
df[df['Yelp'] == '$']
```

```python
def simple(num):
  return num*2
```

```python
simple(2)
```

**Lambda Function**

```python
lambda num: num*2
```

```python
def quality(total_bill, tip):
  if (tip/total_bill) > 0.25:
    return "Generous"
  else:
    return "other"
```

```python
df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
```

```python
df['Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
df
```

**Vectorize method**
It is faster than apply method

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
df.head()
```
