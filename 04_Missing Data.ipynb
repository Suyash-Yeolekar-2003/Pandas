**When reading in missing values, pandas will display it as Nan(Not a number)**


# Options for dealing with Missing Data


1.   Keep it
2.   Remove it
3.   Replace it

**We should know that, there is no 100% correct approach that applies to all circumstances. It depends upon the situation that you encounter**

# Keeping the Missing Data:
**Pros**:

1.   You don't have to do anything. it is as it is.
2.   Data is unchanged or unmanipulated. 
**Cons**

1.   Many methods do not support Nan 
2.   Often there are reasonable guesses.




# Droping or removing the Missing Data
**Pros:**

1.   Easy to do
2.   Based on rules
**Cons:**

1.   Potential to lose a lot of data or useful information.
2.   Limits trained model for future data.

**Droping a row**

*   Makes sense when a lot of info is missing.
*   It is a good point, while droping a row you calculate the percentage of how much data is dropped of the total.

**Droping a Column**

*   Good choice if many rows are missing that particuler feature.










# Filling in the Missing Data
**Pros:**

1.   Potential to save a lot of data for use in training a model

**Cons**

1.   Hardest to do and somewhat arbitrary.
2.   Potential to lead to false conclusions.

**Filling with the same value**

*   Good choice if Nan was a placeholder.

**Filling with interpolated or estimated value**

*   Much harder and require resonable assumptions.
*   List item
